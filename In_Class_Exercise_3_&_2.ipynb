```python
# Starter cell (optional)
import requests
import pandas as pd

API_URL = "https://remoteok.com/api"
```

```python
# Write your solution below.
# Tip: the API response includes a metadata object plus job objects.
# Make sure you keep only real job postings.






import requests
import pandas as pd

API_URL = "https://remoteok.com/api"

# Fetch data
response = requests.get(API_URL)
data = response.json()

# The first item is metadata — remove it
jobs = [job for job in data if isinstance(job, dict) and "id" in job]

# ----------------------------
# Q1: Total number of jobs
# ----------------------------
total_jobs = len(jobs)

# ----------------------------
# Q2: Top 3 most frequent companies
# ----------------------------
companies = [job.get("company", "Unknown") for job in jobs]
top_3_companies = pd.Series(companies).value_counts().head(3)

# ----------------------------
# Q3: Job titles containing "Data" (case-insensitive)
# ----------------------------
data_jobs_count = sum(
    1 for job in jobs
    if "data" in job.get("position", "").lower()
)

# ----------------------------
# Q4: Create DataFrame (first 10 jobs)
# ----------------------------
rows = []

for job in jobs[:10]:
    row = {
        "title": job.get("position"),
        "company": job.get("company"),
        "location": job.get("location", "Unknown") if job.get("location") else "Unknown",
        "salary": job.get("salary") if job.get("salary") else None,
        "tags": ", ".join(job.get("tags", []))
    }
    rows.append(row)

df = pd.DataFrame(rows)

# ----------------------------
# PRINT RESULTS
# ----------------------------
print("Q1: Total job postings =", total_jobs)
print("\nQ2: Top 3 Companies:")
print(top_3_companies)
print("\nQ3: Job titles containing 'Data' =", data_jobs_count)
print("\nQ4: First 10 Jobs DataFrame:")
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)

print(df.head(10))







```

Write a Python program that prompts the user to enter two numbers and perform a division operation. Handle exceptions for both zero division and invalid input (non-numeric input). Display appropriate error messages for each type of exception and ensure the program does not crash due to these errors.

```python
# write your answer here
# Q1: Division with exception handling

try:
    num1 = float(input("Enter first number: "))
    num2 = float(input("Enter second number: "))

    result = num1 / num2
    print("Result:", result)

except ZeroDivisionError:
    print("Error: Cannot divide by zero.")

except ValueError:
    print("Error: Invalid input. Please enter numeric values only.")

except Exception as e:
    print("Unexpected error occurred:", e)

```

 Q2( 4 points)
Define a base class called 'Vehicle' with attributes make and model. Create a derived class Car that inherits from Vehicle and has an additional attribute 'num_doors'. Demonstrate an example of creating an instance of the 'Car' class and accessing its attributes.

```python
# write your answer here
# Q2: Base class Vehicle

class Vehicle:
    def __init__(self, make, model):
        self.make = make
        self.model = model


# Derived class Car
class Car(Vehicle):
    def __init__(self, make, model, num_doors):
        super().__init__(make, model)
        self.num_doors = num_doors


# Create an instance
my_car = Car("Toyota", "Corolla", 4)

# Access attributes
print("Make:", my_car.make)
print("Model:", my_car.model)
print("Number of doors:", my_car.num_doors)

```

Create a program that accepts a list of numbers as input and outputs a new list containing only the even numbers.

```python
# write your answer here

# Q3: Extract even numbers

numbers = list(map(int, input("Enter numbers separated by space: ").split()))

even_numbers = [num for num in numbers if num % 2 == 0]

print("Even numbers:", even_numbers)

```
