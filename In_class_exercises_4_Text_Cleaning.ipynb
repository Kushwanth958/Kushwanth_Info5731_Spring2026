## Q1 (1 point) — Answer below

```python
# Q1 — ANSWER CELL
import pandas as pd

FILE_PATH = "/content/support_messages.txt"

# Load the TAB-separated file into df
df = pd.read_csv(FILE_PATH, sep="\t", names=["id", "message"])

# Print shape and first 3 rows
print("Shape:", df.shape)
print("\nFirst 3 rows:")
print(df.head(3))
```

## Q2 (3 points) — Answer below

```python
# Q2 — ANSWER CELL

# word_count: number of words
df["word_count"] = df["message"].str.split().str.len()

# char_count: number of characters (including spaces)
df["char_count"] = df["message"].str.len()

# num_count: number of digits (0–9)
df["num_count"] = df["message"].str.count(r"\d")

# upper_word_count: number of ALL-CAPS words
df["upper_word_count"] = df["message"].apply(
    lambda text: sum(word.isupper() for word in text.split())
)

# Print full DataFrame
print(df)
```

## Q3 (3 points) — Answer below

```python
# Q3 — ANSWER CELL

import re

# Option A: sklearn stopwords
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
STOPWORDS = set(ENGLISH_STOP_WORDS)

def clean_text(text: str) -> str:
    # 1. Lowercase
    text = text.lower()

    # 2. Remove punctuation/symbols (keep letters, numbers, spaces)
    text = re.sub(r"[^a-z0-9\s]", "", text)

    # 3. Remove stopwords
    tokens = text.split()
    tokens = [word for word in tokens if word not in STOPWORDS]

    # 4. Remove extra spaces
    text = " ".join(tokens)

    return text

# Create new column
df["clean"] = df["message"].apply(clean_text)

# Print original and clean version for id=1 and id=4
for i in [1, 4]:
    row = df[df["id"] == str(i)].iloc[0] # Convert i to string for comparison
    print(f"\nID = {i}")
    print("Original:", row["message"])
    print("Clean   :", row["clean"])
```

## Q4 (2 points) — Answer below

```python
# Q4 — ANSWER CELL

import re

# Extract first occurrence of ORD-#### (case-insensitive)
df["order_id"] = df["message"].str.extract(
    r"(ORD-\d{4})", flags=re.IGNORECASE
)

# Extract first email address (if present)
df["email"] = df["message"].str.extract(
    r"([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})"
)

# Print required columns
print(df[["id", "order_id", "email"]])
```

## Q5 (1 point) — Answer below

```python
# Q5 — ANSWER CELL

from sklearn.feature_extraction.text import TfidfVectorizer
import numpy as np

# Fit TF-IDF on cleaned text
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(df["clean"])

# Compute average TF-IDF score per term
avg_tfidf = np.asarray(X.mean(axis=0)).ravel()

# Get terms
terms = vectorizer.get_feature_names_out()

# Get top 5 highest average TF-IDF terms
top_indices = avg_tfidf.argsort()[-5:][::-1]

print("Top 5 Keywords by Average TF-IDF:\n")
for idx in top_indices:
    print(f"{terms[idx]} : {avg_tfidf[idx]:.4f}")
```
